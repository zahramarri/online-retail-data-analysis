## 7. Missing Values Overview

Assess the amount of missing data in each column before deciding how to handle it.

```python
missing_count = df.isnull().sum()
missing_percentage = (df.isnull().sum() / len(df)) * 100

missing_values = pd.DataFrame({
    "Missing Count": missing_count,
    "Missing Percentage": missing_percentage
})

missing_values
```

### Interpretation

- The `Customer ID` column contains a considerable number of missing values. Its impact will depend on the type of analysis performed later.
- The `Description` column has a small number of missing values, but this column is essential for product-level analysis and will require careful handling during the data cleaning stage.
- No decisions about removing or imputing missing values are made in this notebook.

## 8. Duplicate Rows

Check whether the dataset contains duplicated records before performing any data cleaning.

```python
df.duplicated().sum()

duplicates = df[df.duplicated(keep=False)]
duplicates.sort_values(by=["Invoice", "StockCode"]).head(20)
```

### Interpretation

- The dataset contains **6,865 fully duplicated records**.
- Inspection of duplicated groups confirms that these records are identical across all columns.
- No duplicate records are removed at this stage. The decision will be implemented and documented in the data cleaning notebook.

## 9. Unique Values Overview

Explore the categorical columns to understand their diversity and identify potential data quality issues.

```python
df.nunique()
```

### Interpretation

- The dataset contains transactions from **40 countries**, allowing future geographical analysis.
- The number of unique invoices is much smaller than the number of rows, indicating that each invoice may contain multiple products.
- The number of unique `Description` values is greater than the number of unique `StockCode` values. This suggests that at least some products may have multiple descriptions, requiring further investigation during the exploratory analysis stage.

## Initial Data Quality Assessment

The initial inspection revealed several data quality issues that should be addressed before analysis:

- Missing values in `Customer ID` and `Description`
- Fully duplicated records
- Negative values in the `Price` column
- `Customer ID` stored as float instead of an identifier
- Differences between the number of unique `StockCode` and `Description` values that require further investigation
